**Gráfico de linha**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Gráfico de Linha")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', label='Dados')
plt.legend()  # Adicionando legenda
plt.show()
```

```python
fig, axs = plt.subplots(2)

axs[0].plot(x, y)
axs[0].set_title("Subgráfico 1")

axs[1].bar(x, [3, 5, 7, 9])
axs[1].set_title("Subgráfico 2")

plt.tight_layout()  # Ajustando layout
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
z = [5, 15, 10, 20]

ax.scatter(x, y, z)  # Gráfico de dispersão 3D
plt.show()
```

# **plt.style.use('ggplot')**
O comando plt.style.use('ggplot') é utilizado no Matplotlib para aplicar um estilo visual predefinido aos gráficos. O estilo 'ggplot' em particular é inspirado na biblioteca de visualização de dados R chamada ggplot2, que é famosa por seus gráficos limpos e bem definidos.

O que faz plt.style.use('ggplot')?
Quando você usa plt.style.use('ggplot'), ele aplica um conjunto de configurações visuais ao gráfico, alterando aspectos como:

Cores e Paleta de Cores: A paleta de cores usada nos gráficos se torna mais suave e agradável aos olhos, com cores que imitam o estilo do ggplot2, como tons suaves de cinza e azul.

Fundo Claro e Grade: A grade do gráfico aparece em um fundo claro (geralmente um cinza suave), o que ajuda a visualizar os dados sem distrações.

Linhas e Marcadores: O estilo também ajusta as linhas dos gráficos e os marcadores, tornando-os mais sutis e bem alinhados, de forma a melhorar a legibilidade.

Fontes e Tamanho de Texto: O estilo altera a fonte e o tamanho do texto, fazendo com que títulos, legendas e eixos tenham uma aparência mais uniforme e moderna.

Exemplo de Mudança Visual
Sem o estilo ggplot, um gráfico gerado pelo Matplotlib tem um padrão mais simples, com linhas e cores mais básicas. Ao aplicar plt.style.use('ggplot'), o gráfico fica mais estilizado, com um visual mais polido e com uma aparência mais profissional e limpa, que é o que caracteriza o estilo 'ggplot'.

Exemplo de Antes e Depois:
Sem Estilo 'ggplot':

Cores mais fortes e contrastantes.
O fundo do gráfico é branco.
A grade pode ser mais difícil de visualizar.
Com Estilo 'ggplot':

Paleta de cores mais suave e uniforme.
Fundo cinza suave.
Grade visível, mas de forma discreta, sem competir com os dados.
Esse estilo é bastante útil quando você deseja um visual mais sofisticado e agradável para gráficos, sem ter que ajustar manualmente todos os parâmetros estéticos.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dados de exemplo para o gráfico de dispersão 3D
x = [1, 2, 3, 4, 5]
y = [5, 6, 7, 8, 9]
z = [10, 20, 30, 40, 50]

# Criando a figura e o eixo 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Criando o gráfico de dispersão 3D
ax.scatter(x, y, z)

# Aplicando o estilo 'ggplot'
plt.style.use('ggplot')

# Exibindo o gráfico
plt.show()
```

# plt.annotate

```python
import matplotlib.pyplot as plt

# Dados para o gráfico
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Criando o gráfico de linha
plt.plot(x, y)

# Adicionando uma anotação
plt.annotate('Ponto Alto',
             xy=(x[2], y[2]),  # Posição do ponto a ser anotado
             xytext=(x[2]+0.5, y[2]+5),  # Posição do texto da anotação
             arrowprops=dict(facecolor='black', shrink=0.05))  # Propriedades da seta

# Exibindo o gráfico
plt.title("Exemplo de Anotação")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.show()
```
